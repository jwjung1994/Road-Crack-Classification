```python
import os
import cv2
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from numpy import array
```

## 폴더 디렉토리 설정

```python
TRAIN_DIR = "D:/crack dataset/preprocessing/"
train_folder_list = array(os.listdir(TRAIN_DIR))

train_input = []
train_label = []

label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(train_folder_list)
print(train_folder_list)
print(integer_encoded)   
# negative폴더 = 0(no_crack)
# positive 폴더 = 1(crack)

onehot_encoder = OneHotEncoder(sparse=False) 
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
```

## 폴더마다 데이터 로드 & 할당

```python
for index in range(len(train_folder_list)):
    path = os.path.join(TRAIN_DIR, train_folder_list[index])
    path = path + '/'
    img_list = os.listdir(path)
    for img in img_list:
        img_path = os.path.join(path, img)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        if(img.shape[0] != 128):
            img = cv2.resize(img, dsize = (128, 128), interpolation = cv2.INTER_LINEAR) # (너비, 높이)
        #print(img.shape)
        train_input.append([np.array(img)])
        #train_label.append([np.array(onehot_encoded[index])])          # one_hot 인코딩으로 저장
        train_label.append([np.array(train_folder_list[index])])               # positive, negative로 저장
```

```python
train_input = np.array(train_input)
train_input = train_input.reshape(-1, 128, 128, 3)
train_label = np.reshape(train_label, (-1, 2))

train_input = np.array(train_input).astype(np.float32)
train_label = np.array(train_label).astype(np.float32)
```

```python
print(train_input.shape)
print(train_label.shape)
print(train_label[25000])
```

```python
#npy파일로 저장
np.save("train_data.npy", train_input)
np.save("train_label.npy", train_label)
```
