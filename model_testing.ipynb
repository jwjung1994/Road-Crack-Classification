```python
from tensorflow.keras.models import load_model
import h5py
import numpy as np
```

```python
#model load
model = load_model('trainmodel/0109/trainmodel2.model')
```

```python
# testset load
print("Loading the test data...")
hf = h5py.File('dataset/0110/testset_with_1ch.hdf5', 'r') #Replace the three dots with the directory you saved the dataset in
X = np.array(hf.get('X_concrete'))
y = np.array(hf.get("y_concrete"))
hf.close()
print("Test Data successfully loaded!")
```

```python
loss_and_metrics = model.evaluate(X, y, batch_size=32)

print('')
print('loss : ' + str(loss_and_metrics[0]))
print('accuracy : ' + str(loss_and_metrics[1]))
```

```python
yhat = model.predict_classes(X, verbose = 0)
```

```python
false_positive = 0
false_negative = 0
for i in range(len(yhat)):
    if (yhat[i] != y[i]):
        if(y[i] == 1):          #false_postive
            false_positive += 1
        else:
            false_negative += 1
        print('index : ' + str(i)  +' >>> True : ' + str(y[i]) + ', Predict : ' + str(yhat[i]))
print('거짓 양성 : %d,  거짓 음성 : %d' % (false_positive, false_negative))
```
